```python
from itertools import combinations

def parse_connections(input_data):
    """Parse the input data into a set of connections."""
    connections = set()
    for line in input_data.strip().split('\n'):
        if line:
            a, b = line.split('-')
            connections.add(tuple(sorted([a, b])))
    return connections

def find_connected_triples(connections):
    """Find all sets of three computers that are fully connected to each other."""
    # Get all unique computers
    computers = set()
    for a, b in connections:
        computers.add(a)
        computers.add(b)
    
    triples = []
    # Check all possible combinations of three computers
    for a, b, c in combinations(computers, 3):
        # Check if all three possible connections exist
        if (all(tuple(sorted([x, y])) in connections 
               for x, y in [(a, b), (b, c), (a, c)])):
            triples.append(tuple(sorted([a, b, c])))
    
    return triples

def count_t_triples(triples):
    """Count triples containing at least one computer starting with 't'."""
    return sum(1 for triple in triples 
              if any(comp.startswith('t') for comp in triple))

def solve(input_data):
    connections = parse_connections(input_data)
    triples = find_connected_triples(connections)
    return count_t_triples(triples)

# Test with provided example
test_input = """ur-ug
gd-zg
vc-gp
hc-qt
vf-ni
wa-ij
zy-ch
xd-oa
ex-nc
zr-qf
ys-ls
bd-yp
qt-ph
ld-gj
no-gs
fa-wq
fl-bv
mo-zn
fu-qz
ih-sy
da-uw
pp-am
ok-mz
zo-yg
de-re
pb-ni
bj-np
xa-lt
pu-hj
ui-fx
qq-ak
ah-sh
da-kk
yr-wj
pb-jg
lb-vt
ha-wd
rz-zp
dq-ri
fn-up
mc-zp
tw-vy
gz-cs
yt-hz
sc-zt
gl-pc
kt-ky
id-op
ht-mr
wo-iq
bz-ae
az-dc
so-ou
qd-ua
iy-xn
ch-vi
yq-yi
zc-an
ol-sc
vo-lt
np-va
nh-wf
cv-xu
ij-tl
th-qn
wh-hx
dn-gw
vv-pi
cv-bl
bn-sv
ky-cf
bz-bu
zo-sj
qm-tp
kt-le
cq-an
am-nz
dk-dh
zf-dd
zc-kl
gd-ly
oa-pn
lh-ux
ad-lp
po-kt
mi-xs
dq-ac
tt-la
vo-ax
po-cf
ad-et
qp-op
gw-io
as-mp
eu-ir
mc-br
il-pr
sx-au
mm-qy
yf-jx
bj-oa
sm-wu
cy-pi
dw-xc
mr-iu
dj-az
xk-mc
ri-wc
yj-gw
nr-zf
ru-cq
kn-zo
ey-fc
po-kq
tj-vr
la-ps
pp-mn
rw-ik
ki-jd
rd-xb
vo-os
mh-ss
mw-hx
jg-vf
eg-kk
dg-mo
vf-gh
rt-yq
oa-lw
ls-ef
ry-rc
ll-bc
zc-gr
uk-gd
ra-kl
bu-yl
dk-fs
dk-ph
ch-qp
ft-rx
ic-ze
lm-wx
dv-sh
as-ao
wo-la
qa-mo
kg-eb
id-ii
ng-kt
ws-oo
cp-wt
gj-qq
lj-kz
po-ix
zq-ia
ys-om
yl-zh
zc-fg
yv-qx
zs-ss
wt-ga
vx-dk
wz-cm
zf-ap
ur-ge
le-cf
oa-nu
zw-al
yf-nv
yg-nv
ub-di
ni-fz
gi-dc
az-hh
ow-pq
yx-io
am-hd
yf-lp
ws-fn
xf-qt
od-vf
rj-sk
pb-vf
ar-ir
hs-nt
xx-vt
jz-sj
gd-st
tq-km
tp-sy
ct-wo
bk-ub
tk-ry
pv-qg
ek-bv
zg-at
ie-as
mx-pg
ab-uj
nf-zq
rr-nf
wb-jq
av-da
le-rm
zf-hc
bs-ly
bb-rl
ov-yq
yp-td
hs-gk
gi-cu
zo-xx
sg-eg
qx-bu
rd-yt
dw-sc
bj-ns
ry-qq
de-lb
ww-kd
yf-du
mk-mx
uq-yf
yd-er
qd-vd
mu-mo
xf-nr
bi-bp
vy-jo
to-hn
bv-ga
qg-er
up-bw
rc-wg
wq-sk
up-qu
kh-xa
sl-ql
ck-yw
im-li
fl-ud
la-kb
xf-ph
iu-pc
hh-fa
pb-fz
uq-lz
no-zh
pw-vm
iq-ps
lp-oh
qn-pv
mt-uj
og-nf
zv-fq
xn-ql
yu-bc
da-iz
vp-jq
vm-zv
qu-sl
wf-em
xk-na
ht-wj
wt-bv
ci-vj
lf-gl
st-ub
qf-hv
ou-hd
sv-sr
lo-ct
pz-sx
yw-pr
hd-ih
fu-bm
po-rm
bv-nq
gn-nr
vj-zx
th-fi
gi-hr
od-jg
vm-zh
ub-kw
bp-lt
ef-ul
xw-zg
cp-rn
sl-ob
ky-ng
rt-tw
vy-kv
xs-aq
gr-ru
oa-ln
cw-gk
ar-qy
dp-yi
ea-an
uq-ma
we-nt
sm-km
ab-mx
tk-xq
nh-ys
sb-fb
df-vv
tt-sp
ra-ml
ez-yk
qz-de
hf-mz
lo-iq
px-cm
na-zp
bn-ak
fn-th
hh-hr
np-ww
ow-sl
dr-ic
pb-je
if-zo
xa-pk
ci-bi
zr-ab
dn-yj
of-kb
cm-ub
xn-qx
po-uf
dh-yb
sc-eu
yx-ks
ri-cw
we-wm
rb-br
ud-bv
ju-im
ha-mc
ex-an
vh-gp
dj-dc
sx-rs
dv-zs
ow-xn
tf-hh
ha-xk
fa-dj
of-aw
nq-um
io-bd
us-ft
tw-nl
tk-jk
lo-kb
ga-wb
ox-gh
rt-zv
or-ud
gq-bk
hh-dj
ny-mj
ra-an
bf-dx
oz-ea
qf-vq
bj-xd
xw-bs
wh-cd
vm-yq
gs-ae
in-sb
nh-on
if-jz
hj-yz
qd-ll
gz-ie
ar-mm
up-ui
hj-br
tn-eh
zy-ii
tf-rj
df-dm
ub-ef
uk-vd
px-st
wt-or
ok-qy
xb-ob
tl-mp
fx-dv
wf-tt
pq-ql
uo-ox
lg-qo
zc-ra
jr-yw
cj-oq
de-nv
tn-of
ja-cw
qv-pw
ah-mq
jd-kt
qo-dn
yg-lb
rm-ng
uh-to
cr-uf
cx-xg
jz-zo
ps-np
tw-zx
zs-mq
ej-bh
bi-vy
em-on
ql-yv
no-yb
cy-ce
lc-ur
ou-ie
hr-az
ii-qp
gq-ps
di-bk
wc-cw
mw-cd
fg-id
co-su
aj-zz
ja-jx
rx-qp
ac-qa
fk-xk
mu-dq
ck-jr
cy-zq
rr-km
mm-ok
lz-oh
ci-aa
fa-rj
sw-ab
tb-bm
ng-cf
jd-ew
zf-ph
co-qw
zc-ya
vv-uj
kv-nl
lh-zh
rc-sf
gr-ya
vc-ak
sj-xx
ve-tw
dk-ug
ze-wh
xp-kg
kx-sh
zz-db
zq-df
xg-jq
xc-bf
bd-qo
uq-jx
we-mw
vj-bi
zz-im
pg-xw
dt-aq
fu-to
jf-st
pp-ed
mz-eb
qy-bh
cs-xm
bs-mx
ll-zt
ug-qc
lp-ja
di-uc
yd-lp
ls-em
ng-kq
ju-db
hf-qy
sc-yu
dp-kz
dm-xp
nh-po
sw-hv
hv-au
wo-tn
wu-tq
lk-fm
ur-wa
kn-ut
yp-yj
fy-ir
hn-ds
fs-ny
jy-av
bh-iz
bm-to
ao-mp
hk-kh
rn-um
xz-ca
nv-zo
fx-wn
ct-iq
ob-qu
sk-dj
xt-zt
et-ja
dp-yq
pk-mh
yi-oe
sb-zp
tf-fa
le-ky
ot-ws
ws-mq
mq-sh
cg-pc
qf-sx
wg-jk
sg-cr
uh-wp
ot-mq
ww-ln
ki-cf
jq-as
mm-jd
mt-rs
ya-cq
kz-fq
pp-sb
li-hj
kb-tn
oo-mn
wm-wa
tg-wb
yx-bd
yv-vp
bi-kv
kr-iw
sb-mg
km-uw
bk-hi
nz-ed
jy-uw
yj-qo
yz-rb
an-lm
ps-zw
ra-ex
gp-ez
lf-py
zw-eh
sp-ey
qa-bl
da-cr
ju-li
rp-kt
oc-bz
sv-oj
jx-lz
su-im
be-mo
bb-th
ru-kl
pq-gh
ml-wk
oq-jr
fp-xl
zc-cq
dd-nr
xb-ql
fp-zr
vq-sw
ps-wo
pg-at
uf-rm
qm-rw
kq-uf
nf-ce
eb-hf
gx-hd
nv-re
fb-mc
qm-lk
fk-wd
nj-nc
lb-xx
ss-vo
mo-pd
yp-dn
ro-jl
ac-cv
mx-gd
ap-nr
gp-sv
lg-yx
kg-ce
bc-dx
az-wq
vv-nf
ng-po
th-er
kh-vo
um-wt
lw-ye
ry-sf
zt-qd
jk-qq
qf-tj
zn-xu
eg-jy
kv-dn
ol-do
gr-ex
xt-mr
or-nq
fs-wi
qv-wx
gn-qt
gw-ik
kr-vo
zj-cg
vq-uj
kl-an
vh-xl
dv-mq
fc-em
tu-la
ya-ru
jo-bi
vd-bs
vy-zx
qx-qu
of-al
qx-iy
fk-fb
zw-kb
wj-fu
pg-vd
vx-lc
jl-ie
dv-bw
nu-gp
rt-hz
hf-eu
xt-zj
dr-ij
xz-qz
la-ct
sk-hg
bk-st
di-st
jy-kk
ef-on
pz-au
yk-sr
ou-xm
pr-am
zy-im
px-ub
qm-th
qn-rl
xg-mp
cu-hh
vv-pl
kh-kr
nj-rn
yx-ov
qz-xs
zv-lj
qc-lc
dj-hr
au-rs
et-en
nu-ak
fp-yk
vy-qw
tp-ed
bz-zh
xk-dz
uc-hi
ys-tt
wz-gq
sb-bd
sy-mn
di-hi
bs-zg
fo-yj
eu-mz
qc-iw
wi-ny
fz-vf
iz-ir
lj-yq
wk-kl
hz-ou
ix-rm
yh-cd
uk-mk
ge-dh
lc-fs
nr-wn
db-su
qn-er
ni-ox
ux-ae
ec-rb
sp-em
ov-io
if-yg
nf-cy
gs-ji
ju-hj
kq-ix
fu-uh
eu-fy
mt-pz
ft-mp
ez-nu
hv-rs
qv-fq
ph-bg
zo-ut
fp-sv
ct-zw
ob-yv
ct-kb
pu-lt
vd-at
tg-as
bd-ov
ed-oe
rz-dz
ez-ak
uh-ds
ht-xt
tw-bp
gk-ck
td-hd
wp-xs
wu-cr
xw-uk
ek-wt
fu-hn
zr-vq
wd-dz
qz-to
en-ja
iy-pq
ni-uo
ju-xf
xf-dd
ad-yf
ih-nz
jq-oz
qd-yu
gs-zh
sh-ui
vy-ka
wg-xq
xc-dx
nc-kl
tl-we
wc-hs
uc-zj
df-cy
iq-zw
ur-vx
th-lk
hx-tl
tt-em
yi-qv
ga-nq
sv-nu
vt-if
oq-ri
ut-if
on-xd
az-cu
qy-fy
ih-oe
hf-fy
dh-lc
ws-ui
hc-bg
rs-uj
wg-vr
do-sc
pl-ia
cm-uc
yt-ca
jz-ut
cj-pr
sl-iw
pc-mr
pu-vo
tm-ld
ye-ns
cg-mr
ll-ua
na-sb
sm-ro
ro-so
fs-dh
kn-re
vp-be
on-ul
zx-bp
ac-dg
bc-bf
ot-mj
uo-jg
ey-ul
be-ac
st-cm
tl-wa
pc-py
kg-df
zs-bw
ff-zm
cn-ud
ez-bn
qy-eu
fb-xk
ll-sc
at-ly
wp-fl
um-bv
yi-vm
zj-iu
iz-mz
am-oo
jr-cj
oc-uh
wt-ts
hg-yl
ut-yg
xz-ro
xd-va
co-aj
wu-bj
va-eu
on-sp
ly-xw
nr-qt
cm-jf
gk-jr
zw-lo
id-ki
zv-kz
jq-ao
uw-tq
je-zd
yo-zg
no-oc
yr-py
ma-ad
eb-bh
sr-ez
rx-zy
xu-mu
jr-cw
ux-ra
be-zn
zm-je
zk-vi
xm-ie
ul-nh
wj-gl
vp-qx
ks-dn
rp-jd
vt-nv
zr-lu
lc-ug
xw-xq
yz-ec
sb-ha
lg-kh
am-td
fb-wd
mr-ys
ce-zq
og-df
ii-vi
gr-nj
yo-iu
cx-hp
vx-dh
mu-qj
au-gl
ur-wi
iz-qy
dc-tf
xq-qq
gx-td
ej-fy
yg-cx
ra-ya
ze-wa
bs-yo
gx-nz
ej-gj
pb-gh
fe-yl
kx-fn
rr-wu
ns-ew
kt-rm
ii-sy
df-nf
fk-dz
pc-zj
yl-tx
fl-rn
tu-mr
rm-cf
qz-ds
tw-aa
mm-eu
st-kw
kk-av
ey-om
rb-hj
ui-ot
py-ht
oc-lh
wb-hp
sm-jy
ci-bb
jz-rl
lw-ww
vy-ci
ad-pl
so-ie
hh-gi
ws-kx
cx-kj
ii-ch
zf-bg
kx-ui
tp-gx
ll-do
ug-ny
cf-kq
dr-mw
im-co
kg-cy
xz-ie
lo-al
bm-mi
oz-hp
kz-pw
yh-ze
ld-rc
dz-mg
wf-sp
de-ut
ot-fx
tq-sg
as-cx
mi-jo
bz-ux
px-vq
br-yz
re-sj
hx-yh
yd-yf
kx-up
al-la
ce-df
pv-lk
ts-ud
ru-nj
uq-wn
hk-mh
zo-ek
sg-da
bl-qj
ap-ph
sv-yk
hx-ic
uh-hn
ob-pa
pa-qu
ii-ft
nr-pd
na-fb
bp-ci
yr-pc
ef-fc
ni-zd
lw-pn
pw-yi
vv-kg
kz-yq
wj-xt
ve-aa
bm-dt
yq-fq
rs-sg
wt-fl
fo-ks
kd-bj
fe-lh
ex-ru
if-de
lg-dn
cq-ml
cs-xz
fl-ts
kx-ah
cp-ek
ml-kl
dq-qj
mz-ej
pu-os
vc-nu
rk-hi
xq-jk
gl-yr
zr-sx
lj-su
qx-pq
oa-kd
bf-dw
lm-vm
ps-kb
oc-oz
co-yj
jx-ma
im-rb
au-zr
kx-fx
sx-sw
mp-cx
jr-wc
gn-ww
pg-mk
ru-an
tg-ao
ul-pz
wz-uc
mu-zn
ia-nf
iy-vp
vi-zy
vv-ia
ox-zm
hk-xa
lb-nv
ft-ch
rb-su
wc-nt
um-cn
sv-ak
ew-ln
wi-io
yl-ux
dh-wi
pl-nf
ac-zn
no-lh
op-pe
wk-ya
rs-zr
km-eg
hx-wm
ft-id
ih-ed
vt-de
st-rk
qg-aw
mc-sb
rd-ie
rj-pa
tp-oe
ef-tt
lz-ad
gk-cj
eb-fy
oh-uq
ka-wb
rx-zk
qc-ur
ix-ki
aq-ds
lg-ks
bh-mz
br-su
jl-gz
py-cg
ps-tn
fx-mq
sx-kd
ly-mk
am-ed
nq-ud
tt-ey
yu-xc
xq-gj
np-xd
dk-wi
ew-kd
jg-ni
ov-dn
lt-kh
rp-ky
cn-ek
lb-kn
oh-yd
qw-kv
lh-bu
ge-fs
wq-dc
fz-gi
jx-lp
tt-on
ir-mz
fy-mm
ct-al
dr-wa
pq-xb
vh-nu
og-ia
oe-gx
zx-qw
xz-so
dr-we
iq-la
bj-pn
vq-pz
bi-yo
ut-vt
uk-bs
ej-hf
ya-an
ki-uf
dn-yx
tl-wh
of-wo
ks-qo
mr-zj
bm-uh
cq-ra
xw-sa
ce-vv
ps-of
bz-fe
ju-yz
wf-ef
hn-xs
bf-do
yu-do
hr-cu
av-uw
ia-dm
gk-wc
wx-lj
lz-en
rr-kk
ka-ea
pz-qf
fp-oj
mh-vo
xf-hc
sf-qq
mz-ar
oo-gx
xl-bn
ok-hf
de-yg
to-xs
yf-et
gn-bg
sw-lu
ut-nv
ir-ok
cy-xp
gq-px
sg-wu
eg-da
bi-aa
xc-ua
fs-yb
op-us
gn-xf
eb-ok
hl-nr
ra-gr
hg-gi
vi-fm
kn-if
yu-ol
vx-mj
xc-ol
pb-zd
mt-hv
xq-vr
lh-ae
hf-iz
qo-io
gk-nt
xa-pu
fg-qp
pg-sa
bk-rk
av-sm
rp-xc
qv-yq
fi-uh
gn-dd
gd-bs
qg-bb
ej-qy
xg-hp
lg-io
oq-yw
rk-uc
wq-aq
lc-yb
id-qp
ik-io
bk-wz
ly-zg
pq-sl
qy-eb
rj-cu
ja-lz
pl-zq
da-wu
wp-fu
wd-mg
di-pn
ds-op
sp-om
gh-ni
ve-kv
kb-iq
sr-wz
le-po
wp-qz
om-fc
wq-gp
hp-zv
ki-kq
kx-bw
ob-xn
xm-yt
hd-ed
xn-vp
hl-qt
gw-qo
xb-qx
br-li
qy-ir
ot-zs
dt-qz
yl-ww
sj-db
ze-tl
dq-cv
fk-ha
th-pv
tm-rc
nh-ef
gq-uc
er-fm
ka-as
ma-yd
ve-vj
rx-ii
qx-sl
bd-fo
rl-pv
zz-ju
rw-th
rt-vm
mz-qy
et-jx
tq-sm
pk-kr
rl-fi
xl-sv
zw-tn
df-xp
id-zk
gk-yw
lw-bj
zk-zy
kl-ya
ly-qy
nc-cq
wq-cu
cu-tf
wf-ul
bm-ds
xz-ou
ol-dw
ec-db
lb-ut
cr-kb
as-wb
pp-oe
tg-cx
vd-sa
hi-gq
ob-pq
kv-vj
lu-uj
eb-mm
ze-hx
ew-lw
qd-xc
mh-kh
ux-oc
yb-ur
lw-kd
ys-em
bw-mq
mu-cv
yb-dk
ry-tm
zx-jo
hh-rj
zd-uo
cy-vv
wc-yw
oq-hs
kq-kt
gr-re
ej-iz
vc-sv
am-tp
fx-ws
td-pp
uj-sw
rl-qm
ak-vh
mj-yb
wt-ud
mn-hd
of-iq
wm-ze
kz-rt
tn-ct
ll-dx
pl-df
ie-ro
mi-wp
fx-sh
wt-rn
cq-kl
bu-tx
ya-ow
kt-uf
ny-yb
np-ew
xx-de
dr-cd
jd-ix
km-wu
lw-yr
sw-zr
bz-tx
op-ft
fb-zp
qc-ge
nq-cn
xl-yk
dx-uw
ri-il
nz-mn
jl-cq
ws-dv
yj-ks
os-kr
ix-ng
yp-yx
eu-ar
sa-at
hs-cw
ro-hz
tt-fc
tm-gj
re-if
ri-hs
ds-xs
ej-eu
dt-wp
lm-qv
ab-rs
ye-oa
mt-zr
jl-cs
cj-nt
kk-km
wq-rj
nh-ey
yj-ov
je-gh
lz-et
bc-dw
ed-gx
ax-ss
hj-aj
qj-xu
nl-aa
rb-ju
kd-np
mh-os
wm-mw
wa-ic
mu-bl
sa-gd
ws-ah
kb-wo
bk-jf
en-ad
ka-kj
eu-bh
fz-uo
dg-be
yd-lz
yt-ou
wi-ge
tm-jk
ap-gn
eh-iq
pn-np
wq-gi
ek-or
aq-bm
ga-ud
qg-th
um-fe
gz-ro
qd-dx
or-ga
hv-sx
ky-rm
sa-mk
pb-od
gq-di
aq-tb
fp-ez
aj-br
ht-lf
nt-ck
ao-kj
ab-qf
bg-nr
ny-vx
pi-ce
gq-jf
iw-ss
ui-bw
bb-lk
zs-ui
gd-yo
gj-tj
ye-kd
mk-bs
kj-as
xk-zp
nq-rn
cu-dc
in-zp
yt-cs
ec-zz
ox-hl
tm-qq
sh-zs
ff-uo
dp-vm
wm-cd
eu-ok
va-ln
mt-ab
ec-aj
py-iu
fb-xu
yl-no
cr-tq
ye-pi
ug-vx
ul-om
uk-at
bg-qt
fi-pv
su-li
dc-zn
cp-fl
xs-fu
bf-zt
bw-ot
ys-ef
im-br
ya-nj
oe-oo
je-ox
jq-tg
wo-eh
fm-rw
zr-uj
um-cp
zk-ft
ye-xd
bn-vh
yw-nt
rp-ki
mm-mz
xb-sl
pv-aw
jz-gw
pk-pu
dk-ur
iy-yv
ar-hf
bg-dd
qq-tk
fp-nu
co-ju
zj-gl
wg-tk
ud-um
vq-ab
yh-ij
cj-il
bg-xf
jy-kw
tg-ea
pa-ql
pl-pi
tq-jy
il-gk
ac-bl
ph-hl
cj-ck
nh-sp
eg-tq
kk-cr
gx-am
zw-bc
fq-pw
oj-gp
zo-de
og-zq
kr-ss
ri-cj
bl-dq
tx-ae
zt-dw
xd-ww
oo-ed
vt-sj
fu-tb
av-rr
gs-yl
qw-aa
ot-sh
ra-ru
os-hk
zt-ua
nt-pr
ap-dd
rb-aj
ea-jq
cp-nq
ll-dw
xq-ry
kg-og
tp-mn
jr-hs
st-gq
ys-fc
dx-do
iq-tu
zs-zd
so-xm
pc-lf
dk-qc
zm-vf
du-lp
ez-vh
qm-qg
qu-xn
zx-nl
mj-ur
au-vq
yd-jx
ha-dz
kv-ci
bn-fp
zg-eg
pg-bs
fg-pe
zy-fg
wh-we
bb-aw
vp-pq
fs-qc
pc-ir
sm-rr
ih-am
en-ma
zn-cv
er-aw
ug-fs
ih-zx
ro-xm
ab-pz
oc-ae
gj-tk
mg-ve
lt-iw
va-ew
od-uo
bd-gw
wo-tu
dw-qd
wx-fq
nu-bn
ea-mp
ax-mh
xc-sc
pl-kg
oj-nu
sl-iy
zr-pz
ex-wk
bf-ol
cj-wc
li-mq
ax-pu
hg-hr
bl-mo
xb-iy
hx-we
lu-iq
dg-cv
ff-jg
fb-dz
xa-mh
rz-mg
hl-ap
ai-mr
we-ij
ld-jk
pi-nf
br-db
nz-td
ob-qt
qm-aw
oz-as
ka-xg
dj-gi
yk-nu
al-xl
ny-ge
vq-hv
jx-en
ru-bk
sv-cg
mm-bh
qp-us
yw-hs
hk-iw
kg-pi
tw-qw
hs-il
pk-lf
ml-ya
sy-ed
fb-mg
ht-ai
tj-wg
yr-cg
lf-iu
sc-dx
ea-hp
cq-ex
wo-al
xl-oj
dn-io
lo-wo
mu-pd
tl-wm
zw-tu
mj-wi
rk-lj
iw-pu
sm-sg
rs-pz
dg-mu
fg-rx
sk-hh
mi-ds
ao-hp
ih-pp
ma-du
qt-zf
ns-ww
no-bz
qw-jo
lf-xt
qg-rw
ak-fp
iz-mm
rn-ts
ky-ix
mk-xw
pk-os
gz-xm
rt-fq
yr-ht
xu-pd
fo-dn
pd-cv
zk-pe
pw-wx
xt-pc
yz-db
pg-ly
ri-ck
wb-ao
pz-ks
ae-zh
kh-os
nq-us
en-ci
wz-st
ac-pd
ns-np
fl-nq
oo-hd
id-rx
ok-iz
bf-ua
ln-xd
bb-er
px-jf
qg-fm
uk-ly
sx-mt
pr-cw
wh-wa
cs-ou
vm-kz
wx-zv
gx-sy
hz-ca
sa-bs
fl-um
rm-vh
dx-ol
zh-ji
ax-os
tb-uh
ju-ec
fu-mi
sx-vq
fn-ot
il-wc
qf-rs
eb-ar
ue-hl
lw-ns
ol-zt
pn-va
iq-al
qo-yp
ts-cn
mp-jq
bz-ji
cn-bv
fz-ff
ka-hp
wx-rt
ze-uq
bb-qn
be-dq
sh-up
tx-bg
fc-on
lb-sj
fx-up
yu-dx
rz-mc
ar-fy
pq-qu
np-oa
sf-tj
rk-gq
xq-sf
qa-be
nu-xl
qc-dh
tu-of
tt-ls
oj-vh
tn-xp
vi-qp
yp-ik
om-ef
mo-ac
ae-yl
oa-va
rw-pv
wx-vm
zm-od
sl-vp
lz-br
rn-ud
ez-vc
ri-gk
ri-pr
od-ni
gw-yp
ny-qc
so-hz
gn-ph
sy-oo
lm-pw
jl-ca
pv-bb
er-rl
pk-ss
in-na
gl-ai
rb-co
kl-ex
yh-mw
na-dd
lm-fq
tp-nz
dg-qj
mp-wb
kw-rk
dt-mi
yw-cw
oo-ih
hg-fa
fo-ik
qq-ld
ld-sf
ul-tt
ri-jr
at-em
na-mg
wc-fo
tf-hg
lo-eh
cs-ae
ue-yh
fa-hr
nh-tt
ii-fg
lw-ln
ax-ic
zw-wo
de-pw
wf-ys
tn-lo
bc-do
jx-ad
id-vi
zp-mg
uw-eg
ha-na
xu-dq
sr-fp
ml-gr
lu-ab
bu-ae
lz-yf
rm-ki
bh-hf
fc-wf
uk-mx
id-zy
gh-uo
kh-iw
dq-dg
nf-kg
oq-gk
ka-cx
dm-pl
zz-mn
pn-ln
iz-ar
di-cm
cv-mo
pz-sw
tf-fq
vo-iw
dq-mo
yb-wi
pd-bl
pr-hs
nl-zk
yo-at
vt-re
yb-vx
vv-dm
hg-dj
ew-xd
en-yf
fz-jg
sk-dc
oa-ns
bm-hn
nc-an
lu-rs
gh-zd
ou-jl
rj-gi
xn-pa
az-gi
ou-ro
uq-ad
zv-lm
wh-dr
vt-yg
xk-in
ja-oh
qp-ft
fk-mc
pr-gk
fo-yp
ys-rc
ld-eh
zn-qj
ru-zc
od-zd
gp-sr
ck-pr
fn-zs
ql-iy
fi-qg
lf-wj
su-hj
dt-uh
xx-rz
vr-tm
mc-na
bp-jo
xq-tj
yo-ly
iw-xa
mi-qz
cs-hz
pi-dm
ap-qt
ox-pb
zc-nj
nz-oe
we-cd
yi-wx
vi-dp
lf-yr
ry-wg
jz-lb
pr-jr
fl-cn
wj-py
hp-mp
us-zk
pn-kd
fo-lg
je-kj
pe-ft
ce-ia
ui-fn
fi-qm
jo-ve
gn-wn
or-ts
kv-aa
wx-dp
ck-yv
bw-yx
zd-jg
db-im
ey-em
lk-aw
wx-yq
pe-us
kz-wx
aj-db
ff-pb
ew-ye
cg-wj
ub-uc
wd-sh
pc-wj
rz-fb
if-sj
fk-zp
qg-qn
gz-hz
jf-ng
rc-xq
ye-ww
yu-ln
br-zz
ad-du
tb-hn
hj-zz
rl-vo
mp-oz
ve-bi
cd-ze
wd-mc
aq-hn
fm-pv
pw-dp
as-ea
hx-cd
pc-ai
rw-er
en-lp
rx-us
ey-ls
wp-bm
px-wz
oj-ez
ns-pn
kx-mq
th-aw
bs-vc
ro-rd
tf-dj
to-dt
ug-xn
ov-fo
wk-an
pu-mh
yd-en
ow-pa
lc-wi
yi-kz
wg-sf
tb-to
pq-pa
po-rp
dt-fu
rj-hr
gx-pp
wj-mr
lc-mj
qn-qm
fe-ji
ow-ob
pu-hk
kn-aa
lp-ma
yo-xw
zj-py
fe-gs
sr-vh
yv-xb
yi-lm
je-uo
ai-cg
in-fb
fe-ux
ux-zh
ah-fx
en-du
mg-in
pg-yo
of-ct
wt-ny
sk-ca
pa-vp
yp-ks
zz-li
bb-fi
xx-ut
di-jf
ht-zj
zm-gh
sj-yg
cy-ia
pe-vi
vc-bn
hg-hh
vq-rs
bn-sr
dc-hr
nl-ci
tk-rc
lo-xg
bn-xb
gz-so
cv-qa
xf-hl
lb-zo
ej-eb
hd-pp
yh-tl
ac-qj
gl-tm
bs-at
qn-rw
xl-ez
vf-uo
ro-cs
yw-dj
ji-oc
fe-ae
yd-du
wx-cn
ws-sh
xb-pa
ck-il
lk-rl
tx-oc
ge-yb
xq-ld
im-hj
ok-gx
ua-yu
dh-mj
ik-dn
mh-kr
sf-tk
yt-qg
co-ec
uq-en
iq-tn
yg-re
tl-dr
cx-oz
lh-qa
mm-ej
tb-ds
bk-cm
lh-ji
li-rb
al-tn
kn-jz
sc-bc
sw-au
bw-fn
ky-uf
pw-lj
ch-rx
sv-ez
ts-um
px-bk
ci-jo
hc-ap
ft-fg
og-cy
hi-kw
dm-ls
tu-al
gr-nc
ch-ik
yt-ie
di-kw
ze-we
qj-qa
nv-sj
le-ng
su-aj
dc-hh
qd-sc
qu-ql
dd-qt
av-ut
pv-fk
eg-wu
nc-ya
ng-rp
zy-pe
vd-xw
ru-nc
hj-db
pk-iw
du-lz
mh-lt
jk-rc
br-ju
ud-ni
xf-ap
xt-ai
lw-xd
qu-vp
zv-yi
hc-hl
ak-sr
cf-uf
xn-yv
xa-zf
ly-mx
ue-ap
us-ii
om-tt
np-ye
ik-ks
ns-xd
zk-fg
vj-qw
jd-ng
uc-jf
ma-yf
rz-na
vf-ff
cy-py
od-pd
tw-so
fs-vx
nj-ex
yv-qu
ro-yt
mc-in
ar-ok
ce-og
cg-ht
dv-fn
do-zt
re-xx
le-ki
mn-oe
le-ix
sy-td
ec-lk
zm-gs
vj-aa
wj-iu
sk-hr
oq-dz
al-kb
mx-xw
gs-lh
qj-be
sp-fc
xf-wn
bc-qd
ic-tl
ll-xc
rk-cm
ws-up
ja-ad
ml-bm
fs-mj
ec-im
ix-cf
ex-fa
gd-at
xg-ea
wh-ij
sf-vr
ch-id
bl-zn
us-vi
sj-kn
ga-rn
ll-bf
rm-jd
gq-ub
py-gl
mn-ed
ue-ph
rl-th
pb-zm
kb-tu
xz-xm
lt-pk
do-dw
ex-ya
pw-rt
oc-fe
yo-mk
ov-gw
xk-wd
vd-mk
cx-jq
oz-xg
nj-cq
nt-ri
gw-ks
fl-or
yk-bn
yh-wh
xg-as
ui-ah
oq-vr
dp-rt
eh-ps
ci-ve
iz-eu
ah-up
oz-ka
fx-bw
ph-hc
kj-jq
zo-re
ch-fg
wd-na
vd-gd
qv-kz
yw-cj
ua-dx
zh-oc
ek-ud
jf-rk
ea-cx
sb-dz
yq-zv
va-ns
lz-ma
bp-vj
cn-rn
wn-ue
wu-uw
bz-gs
va-lw
kr-ck
ql-ob
uq-ja
hj-co
dw-yu
ny-ur
hd-xs
ub-rk
xl-sr
lh-yl
ob-iy
vo-xa
ry-kk
cw-nt
iz-fy
ov-ks
px-di
mt-au
vv-og
va-ye
ah-dv
dt-tb
gn-hl
ak-gp
or-um
vo-hk
la-lo
hc-ue
wk-ra
at-mk
cs-so
mw-ij
ol-ll
po-ki
eg-rr
pe-qp
ca-ou
xm-rd
ir-eb
zk-ii
bh-ir
ri-yw
qj-pd
zj-yr
nc-ml
gx-ih
wu-av
fl-ga
hp-kj
ml-vj
in-fk
pz-lu
ix-uf
bn-oj
fe-zh
kl-gr
vc-oj
ug-ge
gi-sk
mj-qc
dp-lm
us-ch
sa-uk
ka-mp
ph-cu
pk-ax
fa-gi
op-ii
de-sj
et-lp
ao-ea
ed-td
fq-lj
yx-gw
ju-su
yr-ai
zp-ha
dg-xu
ra-nj
be-cv
dz-mc
et-yd
mt-sw
db-co
ae-no
wn-qt
zq-vv
rd-cs
pa-sl
up-dv
cy-pl
wp-tb
kr-pu
bp-ve
fp-vh
yz-aj
fn-mq
li-yz
cf-kt
mk-gd
op-vi
qz-bm
yx-ik
zk-qp
ls-om
ji-ae
yd-ja
qv-rw
sp-ls
hk-ax
ij-ze
om-nh
wm-dr
lb-re
on-ls
hn-qz
du-uq
oj-yk
dw-ua
yi-lj
ja-yf
cq-wk
ur-dh
ha-ft
gl-cg
nv-if
bi-nl
mw-ze
xu-be
tu-tn
ln-ye
mg-fk
li-db
yw-il
mr-lf
kg-zq
ls-ul
la-tn
py-mr
yv-ow
vd-yo
rd-gz
kl-jg
rp-ix
kh-ax
ux-ji
ab-sx
uf-ng
jo-aa
kw-wz
wg-uf
ud-cp
vf-ca
tx-lh
zv-dp
ce-dm
rd-so
mi-aq
bu-oc
mn-ih
ss-os
ow-iy
qx-ow
ue-zf
ve-vy
co-li
oh-en
lf-zj
xg-kj
rb-db
ah-bw
ln-np
hf-ir
fl-ek
yk-ol
pe-if
ja-du
ik-lg
ul-em
or-cn
ph-nr
tj-ns
om-em
rt-qv
nz-oo
rl-fm
eh-ct
eg-av
re-ut
wt-cn
kq-jd
yj-bd
tw-jo
dp-qv
xu-bl
ov-ik
qa-zn
fu-ds
tq-da
bu-no
rr-tq
rj-az
nh-ls
jz-nv
sj-ut
ch-zk
qc-vx
ff-lc
kd-ns
qv-zv
la-zw
xk-jk
cj-cw
hc-wn
kd-xd
di-wz
lf-ai
zv-pw
vx-az
nz-lp
io-fo
hn-dt
lk-qn
ca-so
iu-gl
hv-pz
et-du
jz-vt
bu-ux
fi-fm
wm-yh
au-lu
eh-al
pg-ew
bj-ew
df-dw
hd-nz
lz-lp
sm-kk
wm-wh
ea-kj
bc-xc
ht-pc
bp-vy
gz-ca
ql-qx
zg-vd
yt-so
rl-qg
hh-wh
yt-xz
mn-gx
lo-of
yz-zz
ff-cj
dp-fq
jf-kw
as-hp
ma-ja
iz-eb
ox-ff
wi-vx
xt-gl
bh-ez
ge-sw
ur-fs
hi-wz
mw-ic
pn-ye
yz-im
us-zy
je-ni
ys-on
ka-ao
yh-dr
jf-hi
dw-dx
rp-uf
ft-zy
tp-pp
jr-il
lt-ax
ul-fc
yu-ll
pg-zg
iu-ai
nj-an
pu-kh
tq-ak
np-lw
xa-kr
gh-jg
kk-sg
pi-ia
hg-dc
hv-zr
om-uo
dg-qa
ub-wz
kv-bp
bj-ln
jd-po
kd-va
kh-pk
mu-sa
ic-wh
xp-pi
fy-bh
et-oh
ai-wj
ie-cs
cm-hi
ob-qx
il-cw
qc-wi
ka-jq
dm-kg
ij-cd
ht-iu
uh-xs
gz-rx
ea-wb
td-oo
pe-rx
xk-mg
ky-jd
fx-fn
dd-ph
kv-jo
jl-yt
au-ab
ff-je
br-ec
zf-gn
ry-vr
il-nt
cd-tl
mx-dm
ts-bv
oz-ao
pd-dg
cx-ao
fk-na
gs-ux
ca-ie
vr-qq
zd-ff
sl-yv
gi-tf
qx-pa
ru-ml
xa-ax
mw-ah
sw-rs
nc-ic
ar-jf
us-id
oo-pp
yd-ad
lb-if
aq-qz
bv-or
yo-sa
ia-df
ci-qw
xu-mo
cn-ga
no-ux
ej-ir
sb-wd
pd-be
mt-qf
ji-no
yt-gz
cd-ql
ex-ml
sp-ef
zh-bu
ej-ar
ow-qu
zq-pi
cg-lf
vy-aa
to-ds
iy-pa
wh-mw
yl-bz
or-tf
ec-hj
gk-dh
pg-gd
lk-er
fa-sk
mq-ui
ac-mu
nl-vy
xz-gz
uc-px
io-ks
oh-yf
qm-er
dc-fa
et-ma
mk-ix
to-wp
bd-lg
av-sg
dr-mo
du-hk
sm-eg
fm-aw
vc-yk
bj-ww
bf-qd
cg-xt
qm-pv
ac-jx
ox-vf
hl-dd
tj-rc
aq-wp
ic-ij
fx-zs
xk-rz
pl-ce
fm-bb
uw-cr
rw-aw
do-cm
vm-fq
nz-pp
ks-bd
eh-la
os-xa
tb-ni
wk-qn
bb-qm
oq-nt
dt-rb
ey-ef
mn-am
on-ey
qn-aw
kk-wu
tg-ka
uh-aq
lc-dk
fy-ok
vj-jo
uj-le
fo-gw
zq-dm
hp-jq
wm-ij
gq-cm
wk-zc
hl-wn
ok-ej
iy-qu
wn-dd
pg-uk
gr-cq
sp-ul
og-xp
bg-ue
pv-er
df-pi
mq-up
nf-hi
av-km
ov-qo
hg-az
cp-cn
tk-wz
ua-bc
cr-eg
ca-xm
na-dz
ly-vd
bg-ap
ha-fb
yp-lg
ls-fc
av-tq
la-of
vp-ql
zt-dx
sl-xn
uw-sm
sh-bw
uq-yd
fg-us
cp-qa
km-sg
kj-wb
zp-ma
qv-lj
sm-cr
ec-su
jy-wu
xl-gp
co-yz
ge-vx
ia-cf
yx-qo
ox-zd
ve-nl
ct-tu
ol-ua
be-bl
cf-rp
ix-kt
kk-uw
sk-az
vf-zd
dh-ny
zt-yu
ug-yb
uj-qf
uc-bk
oe-am
tx-zh
op-ch
fg-vi
so-jl
wq-hg
hf-hx
xp-vv
or-sp
tp-oo
wd-zp
mx-sa
zq-sf
xs-dt
tg-hr
lk-qg
ds-dt
sx-uj
of-eh
dd-ue
xk-sb
py-ai
lm-lj
bh-ar
qo-oh
hf-mm
jk-sf
iu-xt
no-fe
bz-lh
yz-og
zc-ml
dg-bl
fi-lk
to-ao
gj-sf
mx-zg
qj-cv
lj-vm
sc-ua
yp-io
wg-tm
wq-dj
sk-cu
wc-oq
in-rz
yx-fo
rd-hz
da-sm
ap-hs
ue-qt
gj-wg
km-cr
zc-ex
sy-oe
tk-vr
oh-ad
yr-xt
zy-qp
kh-ss
xb-xn
kx-ot
ts-ek
pq-xn
vh-yk
sr-vc
cw-ck
kq-ky
cp-bv
ww-oa
ih-td
zx-ve
pl-xp
wc-pr
fs-dg
je-fz
zd-fz
mg-mc
li-ec
tx-ji
op-zk
ah-zs
do-ua
xw-at
kn-de
qw-nl
kr-hk
ov-yp
sx-lu
fc-nh
ia-xp
je-vf
om-wf
ra-nc
xz-jl
wn-ap
xx-nv
yg-xx
cr-rr
oh-ma
dk-mj
mt-lu
px-hi
ly-sa
ss-lt
ug-lp
ns-ln
bc-zt
yh-we
le-rp
kn-vt
cj-hs
vt-zo
lg-gw
tf-sk
mz-ul
sg-uw
op-zy
nr-ue
fo-qo
tq-kk
yj-yx
tg-kj
hr-tf
yb-qc
ui-dv
aw-rl
kq-ct
ih-tp
rc-vr
gq-kw
ek-ga
hx-dr
mi-tb
ux-tx
ct-eb
qa-xu
jy-da
jx-oh
ts-nq
pl-og
rd-ou
dh-in
ij-rn
ox-jg
yj-ik
tg-xg
yh-ic
os-bl
uh-mi
kt-ki
wo-uk
vc-vh
jl-xm
ub-hi
mi-to
vr-ld
rw-bb
aj-li
ek-rn
ml-nj
cp-ga
tb-xs
tj-qq
bi-zx
re-jz
pe-ii
bz-cd
lb-ed
st-hi
td-tp
rr-da
zx-ci
pk-hk
fi-er
qf-sw
ge-lc
oc-gs
vy-vj
jy-sg
lt-hk
hz-ie
zp-dz
lu-qf
kw-px
tu-eh
kn-nv
kd-ln
jl-hz
um-ek
bv-rn
kr-ax
qa-mu
do-qd
ni-zm
uj-au
ue-gn
ch-pe
gd-xw
td-oe
or-cp
we-ic
lt-kr
oj-ak
xu-ac
tk-tj
aq-fu
ji-bu
dv-il
oz-kj
au-qf
qn-fm
tg-hp
jl-rd
jy-rr
lm-yq
xa-ss
xx-kn
ki-ng
vo-pk
vc-xl
jx-du
rd-xz
py-xt
um-ga
az-fa
xf-zf
uo-zm
jk-vr
xg-ao
oz-wb
da-km
nf-xp
co-zz
ro-ts
gj-jk
qq-wg
ua-mt
qn-fi
tg-mp
cf-jd
ox-fz
ls-wf
wq-hh
ha-mg
gs-tx
dg-zn
am-sy
cd-wa
oj-sr
fg-op
tw-kv
yv-pq
rt-lm
nc-wk
hg-rj
ff-od
od-je
no-tx
wz-rk
dq-qa
qz-tb
pi-og
bg-hl
ik-bd
zg-sa
ew-oa
jz-xx
zs-kx
rj-dj
wm-ic
yj-io
op-rx
gr-an
wf-on
wn-zf
fn-sh
gp-bn
nu-sr
ry-gj
wt-nq
at-mx
fm-qm
hc-nr
hv-ab
jz-yg
kq-rm
hg-cu
ot-dv
ir-mm
rw-fi
sg-rr
hv-fy
mw-wa
up-zs
zj-ai
oe-mh
cx-wb
rz-ha
rd-ca
tu-ps
th-fm
dn-bd
kl-nj
wk-gr
aj-im
sp-ys
oq-il
xp-ce
ze-dr
ow-ql
kq-rp
lg-yj
cu-dj
jy-km
lu-vq
bv-iy
uq-et
dz-in
pe-id
ld-ry
mn-td
ax-fk
bp-nl
xd-pn
be-mu
kw-bk
fy-mz
wm-pb
ue-xf
ys-ey
dv-kx
mx-vd
ts-ga
du-oh
to-aq
ij-hx
aj-ju
bu-gs
up-ot
ll-fc
iw-os
ow-xb
hv-uj
wk-nj
lg-ov
yv-pa
ck-wc
tj-ld
uf-le
sy-nz
ai-km
sc-bf
gl-ht
ld-tk
mt-vq
nl-jo
rp-rm
wa-we
dk-ge
zm-jg
rb-zz
ps-al
wp-ds
tw-bi
aa-bp
lo-ps
va-ww
qd-ol
ny-lc
mj-ug
bc-ol
jr-nt
zm-fz
kt-ht
fi-aw
cm-kw
zf-hl
on-om
cy-dm
bi-qw
ey-wf
zz-su
ts-cp
uk-yo
nc-zc
fp-gp
zm-zd
ak-yk
iu-cg
xc-zt
sy-hd
qw-ve
lo-tu
mo-qj
gh-ff
xx-if
os-lt
nh-em
st-uc
vc-fp
dq-pd
fz-gh
fn-ah
qq-rc
va-bj
ow-vp
tm-tj
og-dm
hd-tp
cu-fa
gj-vr
hn-wp
kb-eh
bm-xs
wk-ru
jk-bf
hz-xz
lm-kz
ji-yl
wi-ug
kq-le
xl-ey
rj-dc
jf-ub
pb-uo
bu-fe
ny-dk
zg-uk
vj-nl
qv-vm
xc-do
ew-pn
vt-xm
kz-hn
hn-mi
dq-zn
oq-cw
rw-lk
rk-px
ky-kx
aj-jr
jk-ry
cs-ca
vp-ob
lj-dp
yr-iu
bp-qw
tg-oz
dd-hc
zj-wj
wa-hx
hs-ck
qo-hc
xm-hz
qu-xb
od-fz
yg-kn
ge-mj
rt-yi
nq-ek
iw-mh
pr-oq
tl-mw
kw-uc
wd-in
pn-ww
yk-gp
tm-sf
bf-yu
ou-gz
ss-hk
od-ox
mr-yr
zq-xp
zn-pd
tf-az
ah-ot
rz-sb
xg-wb
sv-vh
jg-je
fe-tx
ia-kg
ss-pu
qj-ws
sy-pp
et-ui
xq-tm
ok-bh
ye-bj
hc-gn
cv-kj
fq-yi
po-ky
ry-tj
oj-ce
tw-vj
ji-qp
fk-rz
aa-zx
wf-av
cr-jy
rc-gj
wn-bg
wg-ld
tm-tk
hr-wq
rk-di
lu-hv
rx-vi
wd-rz
od-gh
le-jd
ef-em
yq-pw
ws-bw
of-zw
zg-mk
ha-in
wa-yh
yz-su
ky-ki
rr-uw
kv-zx"""

result = solve(test_input)
print(f"Number of triples containing a computer starting with 't': {result}")
```

```python
from itertools import combinations

def parse_connections(input_data):
    """Parse the input data into a set of connections."""
    connections = set()
    for line in input_data.strip().split('\n'):
        if line:
            a, b = line.split('-')
            connections.add(tuple(sorted([a, b])))
    return connections

def is_fully_connected(computers, connections):
    """Check if all computers in the set are connected to each other."""
    return all(tuple(sorted([a, b])) in connections 
              for a, b in combinations(computers, 2))

def find_largest_clique(connections):
    """Find the largest set of fully connected computers."""
    # Get all unique computers
    computers = set()
    for a, b in connections:
        computers.add(a)
        computers.add(b)
    
    # Try all possible sizes from largest to smallest
    for size in range(len(computers), 1, -1):
        # Check all combinations of that size
        for group in combinations(computers, size):
            if is_fully_connected(group, connections):
                return sorted(group)
    
    return []

def get_lan_password(computers):
    """Get the password by joining computer names with commas."""
    return ','.join(computers)

def solve_part2(input_data):
    connections = parse_connections(input_data)
    largest_group = find_largest_clique(connections)
    return get_lan_password(largest_group)

# Test with provided example
test_input = """ur-ug
gd-zg
vc-gp
hc-qt
vf-ni
wa-ij
zy-ch
xd-oa
ex-nc
zr-qf
ys-ls
bd-yp
qt-ph
ld-gj
no-gs
fa-wq
fl-bv
mo-zn
fu-qz
ih-sy
da-uw
pp-am
ok-mz
zo-yg
de-re
pb-ni
bj-np
xa-lt
pu-hj
ui-fx
qq-ak
ah-sh
da-kk
yr-wj
pb-jg
lb-vt
ha-wd
rz-zp
dq-ri
fn-up
mc-zp
tw-vy
gz-cs
yt-hz
sc-zt
gl-pc
kt-ky
id-op
ht-mr
wo-iq
bz-ae
az-dc
so-ou
qd-ua
iy-xn
ch-vi
yq-yi
zc-an
ol-sc
vo-lt
np-va
nh-wf
cv-xu
ij-tl
th-qn
wh-hx
dn-gw
vv-pi
cv-bl
bn-sv
ky-cf
bz-bu
zo-sj
qm-tp
kt-le
cq-an
am-nz
dk-dh
zf-dd
zc-kl
gd-ly
oa-pn
lh-ux
ad-lp
po-kt
mi-xs
dq-ac
tt-la
vo-ax
po-cf
ad-et
qp-op
gw-io
as-mp
eu-ir
mc-br
il-pr
sx-au
mm-qy
yf-jx
bj-oa
sm-wu
cy-pi
dw-xc
mr-iu
dj-az
xk-mc
ri-wc
yj-gw
nr-zf
ru-cq
kn-zo
ey-fc
po-kq
tj-vr
la-ps
pp-mn
rw-ik
ki-jd
rd-xb
vo-os
mh-ss
mw-hx
jg-vf
eg-kk
dg-mo
vf-gh
rt-yq
oa-lw
ls-ef
ry-rc
ll-bc
zc-gr
uk-gd
ra-kl
bu-yl
dk-fs
dk-ph
ch-qp
ft-rx
ic-ze
lm-wx
dv-sh
as-ao
wo-la
qa-mo
kg-eb
id-ii
ng-kt
ws-oo
cp-wt
gj-qq
lj-kz
po-ix
zq-ia
ys-om
yl-zh
zc-fg
yv-qx
zs-ss
wt-ga
vx-dk
wz-cm
zf-ap
ur-ge
le-cf
oa-nu
zw-al
yf-nv
yg-nv
ub-di
ni-fz
gi-dc
az-hh
ow-pq
yx-io
am-hd
yf-lp
ws-fn
xf-qt
od-vf
rj-sk
pb-vf
ar-ir
hs-nt
xx-vt
jz-sj
gd-st
tq-km
tp-sy
ct-wo
bk-ub
tk-ry
pv-qg
ek-bv
zg-at
ie-as
mx-pg
ab-uj
nf-zq
rr-nf
wb-jq
av-da
le-rm
zf-hc
bs-ly
bb-rl
ov-yq
yp-td
hs-gk
gi-cu
zo-xx
sg-eg
qx-bu
rd-yt
dw-sc
bj-ns
ry-qq
de-lb
ww-kd
yf-du
mk-mx
uq-yf
yd-er
qd-vd
mu-mo
xf-nr
bi-bp
vy-jo
to-hn
bv-ga
qg-er
up-bw
rc-wg
wq-sk
up-qu
kh-xa
sl-ql
ck-yw
im-li
fl-ud
la-kb
xf-ph
iu-pc
hh-fa
pb-fz
uq-lz
no-zh
pw-vm
iq-ps
lp-oh
qn-pv
mt-uj
og-nf
zv-fq
xn-ql
yu-bc
da-iz
vp-jq
vm-zv
qu-sl
wf-em
xk-na
ht-wj
wt-bv
ci-vj
lf-gl
st-ub
qf-hv
ou-hd
sv-sr
lo-ct
pz-sx
yw-pr
hd-ih
fu-bm
po-rm
bv-nq
gn-nr
vj-zx
th-fi
gi-hr
od-jg
vm-zh
ub-kw
bp-lt
ef-ul
xw-zg
cp-rn
sl-ob
ky-ng
rt-tw
vy-kv
xs-aq
gr-ru
oa-ln
cw-gk
ar-qy
dp-yi
ea-an
uq-ma
we-nt
sm-km
ab-mx
tk-xq
nh-ys
sb-fb
df-vv
tt-sp
ra-ml
ez-yk
qz-de
hf-mz
lo-iq
px-cm
na-zp
bn-ak
fn-th
hh-hr
np-ww
ow-sl
dr-ic
pb-je
if-zo
xa-pk
ci-bi
zr-ab
dn-yj
of-kb
cm-ub
xn-qx
po-uf
dh-yb
sc-eu
yx-ks
ri-cw
we-wm
rb-br
ud-bv
ju-im
ha-mc
ex-an
vh-gp
dj-dc
sx-rs
dv-zs
ow-xn
tf-hh
ha-xk
fa-dj
of-aw
nq-um
io-bd
us-ft
tw-nl
tk-jk
lo-kb
ga-wb
ox-gh
rt-zv
or-ud
gq-bk
hh-dj
ny-mj
ra-an
bf-dx
oz-ea
qf-vq
bj-xd
xw-bs
wh-cd
vm-yq
gs-ae
in-sb
nh-on
if-jz
hj-yz
qd-ll
gz-ie
ar-mm
up-ui
hj-br
tn-eh
zy-ii
tf-rj
df-dm
ub-ef
uk-vd
px-st
wt-or
ok-qy
xb-ob
tl-mp
fx-dv
wf-tt
pq-ql
uo-ox
lg-qo
zc-ra
jr-yw
cj-oq
de-nv
tn-of
ja-cw
qv-pw
ah-mq
jd-kt
qo-dn
yg-lb
rm-ng
uh-to
cr-uf
cx-xg
jz-zo
ps-np
tw-zx
zs-mq
ej-bh
bi-vy
em-on
ql-yv
no-yb
cy-ce
lc-ur
ou-ie
hr-az
ii-qp
gq-ps
di-bk
wc-cw
mw-cd
fg-id
co-su
aj-zz
ja-jx
rx-qp
ac-qa
fk-xk
mu-dq
ck-jr
cy-zq
rr-km
mm-ok
lz-oh
ci-aa
fa-rj
sw-ab
tb-bm
ng-cf
jd-ew
zf-ph
co-qw
zc-ya
vv-uj
kv-nl
lh-zh
rc-sf
gr-ya
vc-ak
sj-xx
ve-tw
dk-ug
ze-wh
xp-kg
kx-sh
zz-db
zq-df
xg-jq
xc-bf
bd-qo
uq-jx
we-mw
vj-bi
zz-im
pg-xw
dt-aq
fu-to
jf-st
pp-ed
mz-eb
qy-bh
cs-xm
bs-mx
ll-zt
ug-qc
lp-ja
di-uc
yd-lp
ls-em
ng-kq
ju-db
hf-qy
sc-yu
dp-kz
dm-xp
nh-po
sw-hv
hv-au
wo-tn
wu-tq
lk-fm
ur-wa
kn-ut
yp-yj
fy-ir
hn-ds
fs-ny
jy-av
bh-iz
bm-to
ao-mp
hk-kh
rn-um
xz-ca
nv-zo
fx-wn
ct-iq
ob-qu
sk-dj
xt-zt
et-ja
dp-yq
pk-mh
yi-oe
sb-zp
tf-fa
le-ky
ot-ws
ws-mq
mq-sh
cg-pc
qf-sx
wg-jk
sg-cr
uh-wp
ot-mq
ww-ln
ki-cf
jq-as
mm-jd
mt-rs
ya-cq
kz-fq
pp-sb
li-hj
kb-tn
oo-mn
wm-wa
tg-wb
yx-bd
yv-vp
bi-kv
kr-iw
sb-mg
km-uw
bk-hi
nz-ed
jy-uw
yj-qo
yz-rb
an-lm
ps-zw
ra-ex
gp-ez
lf-py
zw-eh
sp-ey
qa-bl
da-cr
ju-li
rp-kt
oc-bz
sv-oj
jx-lz
su-im
be-mo
bb-th
ru-kl
pq-gh
ml-wk
oq-jr
fp-xl
zc-cq
dd-nr
xb-ql
fp-zr
vq-sw
ps-wo
pg-at
uf-rm
qm-rw
kq-uf
nf-ce
eb-hf
gx-hd
nv-re
fb-mc
qm-lk
fk-wd
nj-nc
lb-xx
ss-vo
mo-pd
yp-dn
ro-jl
ac-cv
mx-gd
ap-nr
gp-sv
lg-yx
kg-ce
bc-dx
az-wq
vv-nf
ng-po
th-er
kh-vo
um-wt
lw-ye
ry-sf
zt-qd
jk-qq
qf-tj
zn-xu
eg-jy
kv-dn
ol-do
gr-ex
xt-mr
or-nq
fs-wi
qv-wx
gn-qt
gw-ik
kr-vo
zj-cg
vq-uj
kl-an
vh-xl
dv-mq
fc-em
tu-la
ya-ru
jo-bi
vd-bs
vy-zx
qx-qu
of-al
qx-iy
fk-fb
zw-kb
wj-fu
pg-vd
vx-lc
jl-ie
dv-bw
nu-gp
rt-hz
hf-eu
xt-zj
dr-ij
xz-qz
la-ct
sk-hg
bk-st
di-st
jy-kk
ef-on
pz-au
yk-sr
ou-xm
pr-am
zy-im
px-ub
qm-th
qn-rl
xg-mp
cu-hh
vv-pl
kh-kr
nj-rn
yx-ov
qz-xs
zv-lj
qc-lc
dj-hr
au-rs
et-en
nu-ak
fp-yk
vy-qw
tp-ed
bz-zh
xk-dz
uc-hi
ys-tt
wz-gq
sb-bd
sy-mn
di-hi
bs-zg
fo-yj
eu-mz
qc-iw
wi-ny
fz-vf
iz-ir
lj-yq
wk-kl
hz-ou
ix-rm
yh-cd
uk-mk
ge-dh
lc-fs
nr-wn
db-su
qn-er
ni-ox
ux-ae
ec-rb
sp-em
ov-io
if-yg
nf-cy
gs-ji
ju-hj
kq-ix
fu-uh
eu-fy
mt-pz
ft-mp
ez-nu
hv-rs
qv-fq
ph-bg
zo-ut
fp-sv
ct-zw
ob-yv
ct-kb
pu-lt
vd-at
tg-as
bd-ov
ed-oe
rz-dz
ez-ak
uh-ds
ht-xt
tw-bp
gk-ck
td-hd
wp-xs
wu-cr
xw-uk
ek-wt
fu-hn
zr-vq
wd-dz
qz-to
en-ja
iy-pq
ni-uo
ju-xf
xf-dd
ad-yf
ih-nz
jq-oz
qd-yu
gs-zh
sh-ui
vy-ka
wg-xq
xc-dx
nc-kl
tl-we
wc-hs
uc-zj
df-cy
iq-zw
ur-vx
th-lk
hx-tl
tt-em
yi-qv
ga-nq
sv-nu
vt-if
oq-ri
ut-if
on-xd
az-cu
qy-fy
ih-oe
hf-fy
dh-lc
ws-ui
hc-bg
rs-uj
wg-vr
do-sc
pl-ia
cm-uc
yt-ca
jz-ut
cj-pr
sl-iw
pc-mr
pu-vo
tm-ld
ye-ns
cg-mr
ll-ua
na-sb
sm-ro
ro-so
fs-dh
kn-re
vp-be
on-ul
zx-bp
ac-dg
bc-bf
ot-mj
uo-jg
ey-ul
be-ac
st-cm
tl-wa
pc-py
kg-df
zs-bw
ff-zm
cn-ud
ez-bn
qy-eu
fb-xk
ll-sc
at-ly
wp-fl
um-bv
yi-vm
zj-iu
iz-mz
am-oo
jr-cj
oc-uh
wt-ts
hg-yl
ut-yg
xz-ro
xd-va
co-aj
wu-bj
va-eu
on-sp
ly-xw
nr-qt
cm-jf
gk-jr
zw-lo
id-ki
zv-kz
jq-ao
uw-tq
je-zd
yo-zg
no-oc
yr-py
ma-ad
eb-bh
sr-ez
rx-zy
xu-mu
jr-cw
ux-ra
be-zn
zm-je
zk-vi
xm-ie
ul-nh
wj-gl
vp-qx
ks-dn
rp-jd
vt-nv
zr-lu
lc-ug
xw-xq
yz-ec
sb-ha
lg-kh
am-td
fb-wd
mr-ys
ce-zq
og-df
ii-vi
gr-nj
yo-iu
cx-hp
vx-dh
mu-qj
au-gl
ur-wi
iz-qy
dc-tf
xq-qq
gx-td
ej-fy
yg-cx
ra-ya
ze-wa
bs-yo
gx-nz
ej-gj
pb-gh
fe-yl
kx-fn
rr-wu
ns-ew
kt-rm
ii-sy
df-nf
fk-dz
pc-zj
yl-tx
fl-rn
tu-mr
rm-cf
qz-ds
tw-aa
mm-eu
st-kw
kk-av
ey-om
rb-hj
ui-ot
py-ht
oc-lh
wb-hp
sm-jy
ci-bb
jz-rl
lw-ww
vy-ci
ad-pl
so-ie
hh-gi
ws-kx
cx-kj
ii-ch
zf-bg
kx-ui
tp-gx
ll-do
ug-ny
cf-kq
dr-mw
im-co
kg-cy
xz-ie
lo-al
bm-mi
oz-hp
kz-pw
yh-ze
ld-rc
dz-mg
wf-sp
de-ut
ot-fx
tq-sg
as-cx
mi-jo
bz-ux
px-vq
br-yz
re-sj
hx-yh
yd-yf
kx-up
al-la
ce-df
pv-lk
ts-ud
ru-nj
uq-wn
hk-mh
zo-ek
sg-da
bl-qj
ap-ph
sv-yk
hx-ic
uh-hn
ob-pa
pa-qu
ii-ft
nr-pd
na-fb
bp-ci
yr-pc
ef-fc
ni-zd
lw-pn
pw-yi
vv-kg
kz-yq
wj-xt
ve-aa
bm-dt
yq-fq
rs-sg
wt-fl
fo-ks
kd-bj
fe-lh
ex-ru
if-de
lg-dn
cq-ml
cs-xz
fl-ts
kx-ah
cp-ek
ml-kl
dq-qj
mz-ej
pu-os
vc-nu
rk-hi
xq-jk
gl-yr
zr-sx
lj-su
qx-pq
oa-kd
bf-dw
lm-vm
ps-kb
oc-oz
co-yj
jx-ma
im-rb
au-zr
kx-fx
sx-sw
mp-cx
jr-wc
gn-ww
pg-mk
ru-an
tg-ao
ul-pz
wz-uc
mu-zn
ia-nf
iy-vp
vi-zy
vv-ia
ox-zm
hk-xa
lb-nv
ft-ch
rb-su
wc-nt
um-cn
sv-ak
ew-ln
wi-io
yl-ux
dh-wi
pl-nf
ac-zn
no-lh
op-pe
wk-ya
rs-zr
km-eg
hx-wm
ft-id
ih-ed
vt-de
st-rk
qg-aw
mc-sb
rd-ie
rj-pa
tp-oe
ef-tt
lz-ad
gk-cj
eb-fy
oh-uq
ka-wb
rx-zk
qc-ur
ix-ki
aq-ds
lg-ks
bh-mz
br-su
jl-gz
py-cg
ps-tn
fx-mq
sx-kd
ly-mk
am-ed
nq-ud
tt-ey
yu-xc
xq-gj
np-xd
dk-wi
ew-kd
jg-ni
ov-dn
lt-kh
rp-ky
cn-ek
lb-kn
oh-yd
qw-kv
lh-bu
ge-fs
wq-dc
fz-gi
jx-lp
tt-on
ir-mz
fy-mm
ct-al
dr-wa
pq-xb
vh-nu
og-ia
oe-gx
zx-qw
xz-so
dr-we
iq-la
bj-pn
vq-pz
bi-yo
ut-vt
uk-bs
ej-hf
ya-an
ki-uf
dn-yx
tl-wh
of-wo
ks-qo
mr-zj
bm-uh
cq-ra
xw-sa
ce-vv
ps-of
bz-fe
ju-yz
wf-ef
hn-xs
bf-do
yu-do
hr-cu
av-uw
ia-dm
gk-wc
wx-lj
lz-en
rr-kk
ka-ea
pz-qf
fp-oj
mh-vo
xf-hc
sf-qq
mz-ar
oo-gx
xl-bn
ok-hf
de-yg
to-xs
yf-et
gn-bg
sw-lu
ut-nv
ir-ok
cy-xp
gq-px
sg-wu
eg-da
bi-aa
xc-ua
fs-yb
op-us
gn-xf
eb-ok
hl-nr
ra-gr
hg-gi
vi-fm
kn-if
yu-ol
vx-mj
xc-ol
pb-zd
mt-hv
xq-vr
lh-ae
hf-iz
qo-io
gk-nt
xa-pu
fg-qp
pg-sa
bk-rk
av-sm
rp-xc
qv-yq
fi-uh
gn-dd
gd-bs
qg-bb
ej-qy
xg-hp
lg-io
oq-yw
rk-uc
wq-aq
lc-yb
id-qp
ik-io
bk-wz
ly-zg
pq-sl
qy-eb
rj-cu
ja-lz
pl-zq
da-wu
wp-fu
wd-mg
di-pn
ds-op
sp-om
gh-ni
ve-kv
kb-iq
sr-wz
le-po
wp-qz
om-fc
wq-gp
hp-zv
ki-kq
kx-bw
ob-xn
xm-yt
hd-ed
xn-vp
hl-qt
gw-qo
xb-qx
br-li
qy-ir
ot-zs
dt-qz
yl-ww
sj-db
ze-tl
dq-cv
fk-ha
th-pv
tm-rc
nh-ef
gq-uc
er-fm
ka-as
ma-yd
ve-vj
rx-ii
qx-sl
bd-fo
rl-pv
zz-ju
rw-th
rt-vm
mz-qy
et-jx
tq-sm
pk-kr
rl-fi
xl-sv
zw-tn
df-xp
id-zk
gk-yw
lw-bj
zk-zy
kl-ya
ly-qy
nc-cq
wq-cu
cu-tf
wf-ul
bm-ds
xz-ou
ol-dw
ec-db
lb-ut
cr-kb
as-wb
pp-oe
tg-cx
vd-sa
hi-gq
ob-pq
kv-vj
lu-uj
eb-mm
ze-hx
ew-lw
qd-xc
mh-kh
ux-oc
yb-ur
lw-kd
ys-em
bw-mq
mu-cv
yb-dk
ry-tm
zx-jo
hh-rj
zd-uo
cy-vv
wc-yw
oq-hs
kq-kt
gr-re
ej-iz
vc-sv
am-tp
fx-ws
td-pp
uj-sw
rl-qm
ak-vh
mj-yb
wt-ud
mn-hd
of-iq
wm-ze
kz-rt
tn-ct
ll-dx
pl-df
ie-ro
mi-wp
fx-sh
wt-rn
cq-kl
bu-tx
ya-ow
kt-uf
ny-yb
np-ew
xx-de
dr-cd
jd-ix
km-wu
lw-yr
sw-zr
bz-tx
op-ft
fb-zp
qc-ge
nq-cn
xl-yk
dx-uw
ri-il
nz-mn
jl-cq
ws-dv
yj-ks
os-kr
ix-ng
yp-yx
eu-ar
sa-at
hs-cw
ro-hz
tt-fc
tm-gj
re-if
ri-hs
ds-xs
ej-eu
dt-wp
lm-qv
ab-rs
ye-oa
mt-zr
jl-cs
cj-nt
kk-km
wq-rj
nh-ey
yj-ov
je-gh
lz-et
bc-dw
ed-gx
ax-ss
hj-aj
qj-xu
nl-aa
rb-ju
kd-np
mh-os
wm-mw
wa-ic
mu-bl
sa-gd
ws-ah
kb-wo
bk-jf
en-ad
ka-kj
eu-bh
fz-uo
dg-be
yd-lz
yt-ou
wi-ge
tm-jk
ap-gn
eh-iq
pn-np
wq-gi
ek-or
aq-bm
ga-ud
qg-th
um-fe
gz-ro
qd-dx
or-ga
hv-sx
ky-rm
sa-mk
pb-od
gq-di
aq-tb
fp-ez
aj-br
ht-lf
nt-ck
ao-kj
ab-qf
bg-nr
ny-vx
pi-ce
gq-jf
iw-ss
ui-bw
bb-lk
zs-ui
gd-yo
gj-tj
ye-kd
mk-bs
kj-as
xk-zp
nq-rn
cu-dc
in-zp
yt-cs
ec-zz
ox-hl
tm-qq
sh-zs
ff-uo
dp-vm
wm-cd
eu-ok
va-ln
mt-ab
ec-aj
py-iu
fb-xu
yl-no
cr-tq
ye-pi
ug-vx
ul-om
uk-at
bg-qt
fi-pv
su-li
dc-zn
cp-fl
xs-fu
bf-zt
bw-ot
ys-ef
im-br
ya-nj
oe-oo
je-ox
jq-tg
wo-eh
fm-rw
zr-uj
um-cp
zk-ft
ye-xd
bn-vh
yw-nt
rp-ki
mm-mz
xb-sl
pv-aw
jz-gw
pk-pu
dk-ur
iy-yv
ar-hf
bg-dd
qq-tk
fp-nu
co-ju
zj-gl
wg-tk
ud-um
vq-ab
yh-ij
cj-il
bg-xf
jy-kw
tg-ea
pa-ql
pl-pi
tq-jy
il-gk
ac-bl
ph-hl
cj-ck
nh-sp
eg-tq
kk-cr
gx-am
zw-bc
fq-pw
oj-gp
zo-de
og-zq
kr-ss
ri-cj
bl-dq
tx-ae
zt-dw
xd-ww
oo-ed
vt-sj
fu-tb
av-rr
gs-yl
qw-aa
ot-sh
ra-ru
os-hk
zt-ua
nt-pr
ap-dd
rb-aj
ea-jq
cp-nq
ll-dw
xq-ry
kg-og
tp-mn
jr-hs
st-gq
ys-fc
dx-do
iq-tu
zs-zd
so-xm
pc-lf
dk-qc
zm-vf
du-lp
ez-vh
qm-qg
qu-xn
zx-nl
mj-ur
au-vq
yd-jx
ha-dz
kv-ci
bn-fp
zg-eg
pg-bs
fg-pe
zy-fg
wh-we
bb-aw
vp-pq
fs-qc
pc-ir
sm-rr
ih-am
en-ma
zn-cv
er-aw
ug-fs
ih-zx
ro-xm
ab-pz
oc-ae
gj-tk
mg-ve
lt-iw
va-ew
od-uo
bd-gw
wo-tu
dw-qd
wx-fq
nu-bn
ea-mp
ax-mh
xc-sc
pl-kg
oj-nu
sl-iy
zr-pz
ex-wk
bf-ol
cj-wc
li-mq
ax-pu
hg-hr
bl-mo
xb-iy
hx-we
lu-iq
dg-cv
ff-jg
fb-dz
xa-mh
rz-mg
hl-ap
ai-mr
we-ij
ld-jk
pi-nf
br-db
nz-td
ob-qt
qm-aw
oz-as
ka-xg
dj-gi
yk-nu
al-xl
ny-ge
vq-hv
jx-en
ru-bk
sv-cg
mm-bh
qp-us
yw-hs
hk-iw
kg-pi
tw-qw
hs-il
pk-lf
ml-ya
sy-ed
fb-mg
ht-ai
tj-wg
yr-cg
lf-iu
sc-dx
ea-hp
cq-ex
wo-al
xl-oj
dn-io
lo-wo
mu-pd
tl-wm
zw-tu
mj-wi
rk-lj
iw-pu
sm-sg
rs-pz
dg-mu
fg-rx
sk-hh
mi-ds
ao-hp
ih-pp
ma-du
qt-zf
ns-ww
no-bz
qw-jo
lf-xt
qg-rw
ak-fp
iz-mm
rn-ts
ky-ix
mk-xw
pk-os
gz-xm
rt-fq
yr-ht
xu-pd
fo-dn
pd-cv
zk-pe
pw-wx
xt-pc
yz-db
pg-ly
ri-ck
wb-ao
pz-ks
ae-zh
kh-os
nq-us
en-ci
wz-st
ac-pd
ns-np
fl-nq
oo-hd
id-rx
ok-iz
bf-ua
ln-xd
bb-er
px-jf
qg-fm
uk-ly
sx-mt
pr-cw
wh-wa
cs-ou
vm-kz
wx-zv
gx-sy
hz-ca
sa-bs
fl-um
rm-vh
dx-ol
zh-ji
ax-os
tb-uh
ju-ec
fu-mi
sx-vq
fn-ot
il-wc
qf-rs
eb-ar
ue-hl
lw-ns
ol-zt
pn-va
iq-al
qo-yp
ts-cn
mp-jq
bz-ji
cn-bv
fz-ff
ka-hp
wx-rt
ze-uq
bb-qn
be-dq
sh-up
tx-bg
fc-on
lb-sj
fx-up
yu-dx
rz-mc
ar-fy
pq-qu
np-oa
sf-tj
rk-gq
xq-sf
qa-be
nu-xl
qc-dh
tu-of
tt-ls
oj-vh
tn-xp
vi-qp
yp-ik
om-ef
mo-ac
ae-yl
oa-va
rw-pv
wx-vm
zm-od
sl-vp
lz-br
rn-ud
ez-vc
ri-gk
ri-pr
od-ni
gw-yp
ny-qc
so-hz
gn-ph
sy-oo
lm-pw
jl-ca
pv-bb
er-rl
pk-ss
in-na
gl-ai
rb-co
kl-ex
yh-mw
na-dd
lm-fq
tp-nz
dg-qj
mp-wb
kw-rk
dt-mi
yw-cw
oo-ih
hg-fa
fo-ik
qq-ld
ld-sf
ul-tt
ri-jr
at-em
na-mg
wc-fo
tf-hg
lo-eh
cs-ae
ue-yh
fa-hr
nh-tt
ii-fg
lw-ln
ax-ic
zw-wo
de-pw
wf-ys
tn-lo
bc-do
jx-ad
id-vi
zp-mg
uw-eg
ha-na
xu-dq
sr-fp
ml-gr
lu-ab
bu-ae
lz-yf
rm-ki
bh-hf
fc-wf
uk-mx
id-zy
gh-uo
kh-iw
dq-dg
nf-kg
oq-gk
ka-cx
dm-pl
zz-mn
pn-ln
iz-ar
di-cm
cv-mo
pz-sw
tf-fq
vo-iw
dq-mo
yb-wi
pd-bl
pr-hs
nl-zk
yo-at
vt-re
yb-vx
vv-dm
hg-dj
ew-xd
en-yf
fz-jg
sk-dc
oa-ns
bm-hn
nc-an
lu-rs
gh-zd
ou-jl
rj-gi
xn-pa
az-gi
ou-ro
uq-ad
zv-lm
wh-dr
vt-yg
xk-in
ja-oh
qp-ft
fk-mc
pr-gk
fo-yp
ys-rc
ld-eh
zn-qj
ru-zc
od-zd
gp-sr
ck-pr
fn-zs
ql-iy
fi-qg
lf-wj
su-hj
dt-uh
xx-rz
vr-tm
mc-na
bp-jo
xq-tj
yo-ly
iw-xa
mi-qz
cs-hz
pi-dm
ap-qt
ox-pb
zc-nj
nz-oe
we-cd
yi-wx
vi-dp
lf-yr
ry-wg
jz-lb
pr-jr
fl-cn
wj-py
hp-mp
us-zk
pn-kd
fo-lg
je-kj
pe-ft
ce-ia
ui-fn
fi-qm
jo-ve
gn-wn
or-ts
kv-aa
wx-dp
ck-yv
bw-yx
zd-jg
db-im
ey-em
lk-aw
wx-yq
pe-us
kz-wx
aj-db
ff-pb
ew-ye
cg-wj
ub-uc
wd-sh
pc-wj
rz-fb
if-sj
fk-zp
qg-qn
gz-hz
jf-ng
rc-xq
ye-ww
yu-ln
br-zz
ad-du
tb-hn
hj-zz
rl-vo
mp-oz
ve-bi
cd-ze
wd-mc
aq-hn
fm-pv
pw-dp
as-ea
hx-cd
pc-ai
rw-er
en-lp
rx-us
ey-ls
wp-bm
px-wz
oj-ez
ns-pn
kx-mq
th-aw
bs-vc
ro-rd
tf-dj
to-dt
ug-xn
ov-fo
wk-an
pu-mh
yd-en
ow-pa
lc-wi
yi-kz
wg-sf
tb-to
pq-pa
po-rp
dt-fu
rj-hr
gx-pp
wj-mr
lc-mj
qn-qm
fe-ji
ow-ob
pu-hk
kn-aa
lp-ma
yo-xw
zj-py
fe-gs
sr-vh
yv-xb
yi-lm
je-uo
ai-cg
in-fb
fe-ux
ux-zh
ah-fx
en-du
mg-in
pg-yo
of-ct
wt-ny
sk-ca
pa-vp
yp-ks
zz-li
bb-fi
xx-ut
di-jf
ht-zj
zm-gh
sj-yg
cy-ia
pe-vi
vc-bn
hg-hh
vq-rs
bn-sr
dc-hr
nl-ci
tk-rc
lo-xg
bn-xb
gz-so
cv-qa
xf-hl
lb-zo
ej-eb
hd-pp
yh-tl
ac-qj
gl-tm
bs-at
qn-rw
xl-ez
vf-uo
ro-cs
yw-dj
ji-oc
fe-ae
yd-du
wx-cn
ws-sh
xb-pa
ck-il
lk-rl
tx-oc
ge-yb
xq-ld
im-hj
ok-gx
ua-yu
dh-mj
ik-dn
mh-kr
sf-tk
yt-qg
co-ec
uq-en
iq-tn
yg-re
tl-dr
cx-oz
lh-qa
mm-ej
tb-ds
bk-cm
lh-ji
li-rb
al-tn
kn-jz
sc-bc
sw-au
bw-fn
ky-uf
pw-lj
ch-rx
sv-ez
ts-um
px-bk
ci-jo
hc-ap
ft-fg
og-cy
hi-kw
dm-ls
tu-al
gr-nc
ch-ik
yt-ie
di-kw
ze-we
qj-qa
nv-sj
le-ng
su-aj
dc-hh
qd-sc
qu-ql
dd-qt
av-ut
pv-fk
eg-wu
nc-ya
ng-rp
zy-pe
vd-xw
ru-nc
hj-db
pk-iw
du-lz
mh-lt
jk-rc
br-ju
ud-ni
xf-ap
xt-ai
lw-xd
qu-vp
zv-yi
hc-hl
ak-sr
cf-uf
xn-yv
xa-zf
ly-mx
ue-ap
us-ii
om-tt
np-ye
ik-ks
ns-xd
zk-fg
vj-qw
jd-ng
uc-jf
ma-yf
rz-na
vf-ff
cy-py
od-pd
tw-so
fs-vx
nj-ex
yv-qu
ro-yt
mc-in
ar-ok
ce-og
cg-ht
dv-fn
do-zt
re-xx
le-ki
mn-oe
le-ix
sy-td
ec-lk
zm-gs
vj-aa
wj-iu
sk-hr
oq-dz
al-kb
mx-xw
gs-lh
qj-be
sp-fc
xf-wn
bc-qd
ic-tl
ll-xc
rk-cm
ws-up
ja-ad
ml-bm
fs-mj
ec-im
ix-cf
ex-fa
gd-at
xg-ea
wh-ij
sf-vr
ch-id
bl-zn
us-vi
sj-kn
ga-rn
ll-bf
rm-jd
gq-ub
py-gl
mn-ed
ue-ph
rl-th
pb-zm
kb-tu
xz-xm
lt-pk
do-dw
ex-ya
pw-rt
oc-fe
yo-mk
ov-gw
xk-wd
vd-mk
cx-jq
oz-xg
nj-cq
nt-ri
gw-ks
fl-or
yk-bn
yh-wh
xg-as
ui-ah
oq-vr
dp-rt
eh-ps
ci-ve
iz-eu
ah-up
oz-ka
fx-bw
ph-hc
kj-jq
zo-re
ch-fg
wd-na
vd-gd
qv-kz
yw-cj
ua-dx
zh-oc
ek-ud
jf-rk
ea-cx
sb-dz
yq-zv
va-ns
lz-ma
bp-vj
cn-rn
wn-ue
wu-uw
bz-gs
va-lw
kr-ck
ql-ob
uq-ja
hj-co
dw-yu
ny-ur
hd-xs
ub-rk
xl-sr
lh-yl
ob-iy
vo-xa
ry-kk
cw-nt
iz-fy
ov-ks
px-di
mt-au
vv-og
va-ye
ah-dv
dt-tb
gn-hl
ak-gp
or-um
vo-hk
la-lo
hc-ue
wk-ra
at-mk
cs-so
mw-ij
ol-ll
po-ki
eg-rr
pe-qp
ca-ou
xm-rd
ir-eb
zk-ii
bh-ir
ri-yw
qj-pd
zj-yr
nc-ml
gx-ih
wu-av
fl-ga
hp-kj
ml-vj
in-fk
pz-lu
ix-uf
bn-oj
fe-zh
kl-gr
vc-oj
ug-ge
gi-sk
mj-qc
dp-lm
us-ch
sa-uk
ka-mp
ph-cu
pk-ax
fa-gi
op-ii
de-sj
et-lp
ao-ea
ed-td
fq-lj
yx-gw
ju-su
yr-ai
zp-ha
dg-xu
ra-nj
be-cv
dz-mc
et-yd
mt-sw
db-co
ae-no
wn-qt
zq-vv
rd-cs
pa-sl
up-dv
cy-pl
wp-tb
kr-pu
bp-ve
fp-vh
yz-aj
fn-mq
li-yz
cf-kt
mk-gd
op-vi
qz-bm
yx-ik
zk-qp
ls-om
ji-ae
yd-ja
qv-rw
sp-ls
hk-ax
ij-ze
om-nh
wm-dr
lb-re
on-ls
hn-qz
du-uq
oj-yk
dw-ua
yi-lj
ja-yf
cq-wk
ur-dh
ha-ft
gl-cg
nv-if
bi-nl
mw-ze
xu-be
tu-tn
ln-ye
mg-fk
li-db
yw-il
mr-lf
kg-zq
ls-ul
la-tn
py-mr
yv-ow
vd-yo
rd-gz
kl-jg
rp-ix
kh-ax
ux-ji
ab-sx
uf-ng
jo-aa
kw-wz
wg-uf
ud-cp
vf-ca
tx-lh
zv-dp
ce-dm
rd-so
mi-aq
bu-oc
mn-ih
ss-os
ow-iy
qx-ow
ue-zf
ve-vy
co-li
oh-en
lf-zj
xg-kj
rb-db
ah-bw
ln-np
hf-ir
fl-ek
yk-ol
pe-if
ja-du
ik-lg
ul-em
or-cn
ph-nr
tj-ns
om-em
rt-qv
nz-oo
rl-fm
eh-ct
eg-av
re-ut
wt-cn
kq-jd
yj-bd
tw-jo
dp-qv
xu-bl
ov-ik
qa-zn
fu-ds
tq-da
bu-no
rr-tq
rj-az
nh-ls
jz-nv
sj-ut
ch-zk
qc-vx
ff-lc
kd-ns
qv-zv
la-zw
xk-jk
cj-cw
hc-wn
kd-xd
di-wz
lf-ai
zv-pw
vx-az
nz-lp
io-fo
hn-dt
lk-qn
ca-so
iu-gl
hv-pz
et-du
jz-vt
bu-ux
fi-fm
wm-yh
au-lu
eh-al
pg-ew
bj-ew
df-dw
hd-nz
lz-lp
sm-kk
wm-wh
ea-kj
bc-xc
ht-pc
bp-vy
gz-ca
ql-qx
zg-vd
yt-so
rl-qg
hh-wh
yt-xz
mn-gx
lo-of
yz-zz
ff-cj
dp-fq
jf-kw
as-hp
ma-ja
iz-eb
ox-ff
wi-vx
xt-gl
bh-ez
ge-sw
ur-fs
hi-wz
mw-ic
pn-ye
yz-im
us-zy
je-ni
ys-on
ka-ao
yh-dr
jf-hi
dw-dx
rp-uf
ft-zy
tp-pp
jr-il
lt-ax
ul-fc
yu-ll
pg-zg
iu-ai
nj-an
pu-kh
tq-ak
np-lw
xa-kr
gh-jg
kk-sg
pi-ia
hg-dc
hv-zr
om-uo
dg-qa
ub-wz
kv-bp
bj-ln
jd-po
kd-va
kh-pk
mu-sa
ic-wh
xp-pi
fy-bh
et-oh
ai-wj
ie-cs
cm-hi
ob-qx
il-cw
qc-wi
ka-jq
dm-kg
ij-cd
ht-iu
uh-xs
gz-rx
ea-wb
td-oo
pe-rx
xk-mg
ky-jd
fx-fn
dd-ph
kv-jo
jl-yt
au-ab
ff-je
br-ec
zf-gn
ry-vr
il-nt
cd-tl
mx-dm
ts-bv
oz-ao
pd-dg
cx-ao
fk-na
gs-ux
ca-ie
vr-qq
zd-ff
sl-yv
gi-tf
qx-pa
ru-ml
xa-ax
mw-ah
sw-rs
nc-ic
ar-jf
us-id
oo-pp
yd-ad
lb-if
aq-qz
bv-or
yo-sa
ia-df
ci-qw
xu-mo
cn-ga
no-ux
ej-ir
sb-wd
pd-be
mt-qf
ji-no
yt-gz
cd-ql
ex-ml
sp-ef
zh-bu
ej-ar
ow-qu
zq-pi
cg-lf
vy-aa
to-ds
iy-pa
wh-mw
yl-bz
or-tf
ec-hj
gk-dh
pg-gd
lk-er
fa-sk
mq-ui
ac-mu
nl-vy
xz-gz
uc-px
io-ks
oh-yf
qm-er
dc-fa
et-ma
mk-ix
to-wp
bd-lg
av-sg
dr-mo
du-hk
sm-eg
fm-aw
vc-yk
bj-ww
bf-qd
cg-xt
qm-pv
ac-jx
ox-vf
hl-dd
tj-rc
aq-wp
ic-ij
fx-zs
xk-rz
pl-ce
fm-bb
uw-cr
rw-aw
do-cm
vm-fq
nz-pp
ks-bd
eh-la
os-xa
tb-ni
wk-qn
bb-qm
oq-nt
dt-rb
ey-ef
mn-am
on-ey
qn-aw
kk-wu
tg-ka
uh-aq
lc-dk
fy-ok
vj-jo
uj-le
fo-gw
zq-dm
hp-jq
wm-ij
gq-cm
wk-zc
hl-wn
ok-ej
iy-qu
wn-dd
pg-uk
gr-cq
sp-ul
og-xp
bg-ue
pv-er
df-pi
mq-up
nf-hi
av-km
ov-qo
hg-az
cp-cn
tk-wz
ua-bc
cr-eg
ca-xm
na-dz
ly-vd
bg-ap
ha-fb
yp-lg
ls-fc
av-tq
la-of
vp-ql
zt-dx
sl-xn
uw-sm
sh-bw
uq-yd
fg-us
cp-qa
km-sg
kj-wb
zp-ma
qv-lj
sm-cr
ec-su
jy-wu
xl-gp
co-yz
ge-vx
ia-cf
yx-qo
ox-zd
ve-nl
ct-tu
ol-ua
be-bl
cf-rp
ix-kt
kk-uw
sk-az
vf-zd
dh-ny
zt-yu
ug-yb
uj-qf
uc-bk
oe-am
tx-zh
op-ch
fg-vi
so-jl
wq-hg
hf-hx
xp-vv
or-sp
tp-oo
wd-zp
mx-sa
zq-sf
xs-dt
tg-hr
lk-qg
ds-dt
sx-uj
of-eh
dd-ue
xk-sb
py-ai
lm-lj
bh-ar
qo-oh
hf-mm
jk-sf
iu-xt
no-fe
bz-lh
yz-og
zc-ml
dg-bl
fi-lk
to-ao
gj-sf
mx-zg
qj-cv
lj-vm
sc-ua
yp-io
wg-tm
wq-dj
sk-cu
wc-oq
in-rz
yx-fo
rd-hz
da-sm
ap-hs
ue-qt
gj-wg
km-cr
zc-ex
sy-oe
tk-vr
oh-ad
yr-xt
zy-qp
kh-ss
xb-xn
kx-ot
ts-ek
pq-xn
vh-yk
sr-vc
cw-ck
kq-ky
cp-bv
ww-oa
ih-td
zx-ve
pl-xp
wc-pr
fs-dg
je-fz
zd-fz
mg-mc
li-ec
tx-ji
op-zk
ah-zs
do-ua
xw-at
kn-de
qw-nl
kr-hk
ov-yp
sx-lu
fc-nh
ia-xp
je-vf
om-wf
ra-nc
xz-jl
wn-ap
xx-nv
yg-xx
cr-rr
oh-ma
dk-mj
mt-lu
px-hi
ly-sa
ss-lt
ug-lp
ns-ln
bc-zt
yh-we
le-rp
kn-vt
cj-hs
vt-zo
lg-gw
tf-sk
mz-ul
sg-uw
op-zy
nr-ue
fo-qo
tq-kk
yj-yx
tg-kj
hr-tf
yb-qc
ui-dv
aw-rl
kq-ct
ih-tp
rc-vr
gq-kw
ek-ga
hx-dr
mi-tb
ux-tx
ct-eb
qa-xu
jy-da
jx-oh
ts-nq
pl-og
rd-ou
dh-in
ij-rn
ox-jg
yj-ik
tg-xg
yh-ic
os-bl
uh-mi
kt-ki
wo-uk
vc-vh
jl-xm
ub-hi
mi-to
vr-ld
rw-bb
aj-li
ek-rn
ml-nj
cp-ga
tb-xs
tj-qq
bi-zx
re-jz
pe-ii
bz-cd
lb-ed
st-hi
td-tp
rr-da
zx-ci
pk-hk
fi-er
qf-sw
ge-lc
oc-gs
vy-vj
jy-sg
lt-hk
hz-ie
zp-dz
lu-qf
kw-px
tu-eh
kn-nv
kd-ln
jl-hz
um-ek
bv-rn
kr-ax
qa-mu
do-qd
ni-zm
uj-au
ue-gn
ch-pe
gd-xw
td-oe
or-cp
we-ic
lt-kr
oj-ak
xu-ac
tk-tj
aq-fu
ji-bu
dv-il
oz-kj
au-qf
qn-fm
tg-hp
jl-rd
jy-rr
lm-yq
xa-ss
xx-kn
ki-ng
vo-pk
vc-xl
jx-du
rd-xz
py-xt
um-ga
az-fa
xf-zf
uo-zm
jk-vr
xg-ao
oz-wb
da-km
nf-xp
co-zz
ro-ts
gj-jk
qq-wg
ua-mt
qn-fi
tg-mp
cf-jd
ox-fz
ls-wf
wq-hh
ha-mg
gs-tx
dg-zn
am-sy
cd-wa
oj-sr
fg-op
tw-kv
yv-pq
rt-lm
nc-wk
hg-rj
ff-od
od-je
no-tx
wz-rk
dq-qa
qz-tb
pi-og
bg-hl
ik-bd
zg-sa
ew-oa
jz-xx
zs-kx
rj-dj
wm-ic
yj-io
op-rx
gr-an
wf-on
wn-zf
fn-sh
gp-bn
nu-sr
ry-gj
wt-nq
at-mx
fm-qm
hc-nr
hv-ab
jz-yg
kq-rm
hg-cu
ot-dv
ir-mm
rw-fi
sg-rr
hv-fy
mw-wa
up-zs
zj-ai
oe-mh
cx-wb
rz-ha
rd-ca
tu-ps
th-fm
dn-bd
kl-nj
wk-gr
aj-im
sp-ys
oq-il
xp-ce
ze-dr
ow-ql
kq-rp
lg-yj
cu-dj
jy-km
lu-vq
bv-iy
uq-et
dz-in
pe-id
ld-ry
mn-td
ax-fk
bp-nl
xd-pn
be-mu
kw-bk
fy-mz
wm-pb
ue-xf
ys-ey
dv-kx
mx-vd
ts-ga
du-oh
to-aq
ij-hx
aj-ju
bu-gs
up-ot
ll-fc
iw-os
ow-xb
hv-uj
wk-nj
lg-ov
yv-pa
ck-wc
tj-ld
uf-le
sy-nz
ai-km
sc-bf
gl-ht
ld-tk
mt-vq
nl-jo
rp-rm
wa-we
dk-ge
zm-jg
rb-zz
ps-al
wp-ds
tw-bi
aa-bp
lo-ps
va-ww
qd-ol
ny-lc
mj-ug
bc-ol
jr-nt
zm-fz
kt-ht
fi-aw
cm-kw
zf-hl
on-om
cy-dm
bi-qw
ey-wf
zz-su
ts-cp
uk-yo
nc-zc
fp-gp
zm-zd
ak-yk
iu-cg
xc-zt
sy-hd
qw-ve
lo-tu
mo-qj
gh-ff
xx-if
os-lt
nh-em
st-uc
vc-fp
dq-pd
fz-gh
fn-ah
qq-rc
va-bj
ow-vp
tm-tj
og-dm
hd-tp
cu-fa
gj-vr
hn-wp
kb-eh
bm-xs
wk-ru
jk-bf
hz-xz
lm-kz
ji-yl
wi-ug
kq-le
xl-ey
rj-dc
jf-ub
pb-uo
bu-fe
ny-dk
zg-uk
vj-nl
qv-vm
xc-do
ew-pn
vt-xm
kz-hn
hn-mi
dq-zn
oq-cw
rw-lk
rk-px
ky-kx
aj-jr
jk-ry
cs-ca
vp-ob
lj-dp
yr-iu
bp-qw
tg-oz
dd-hc
zj-wj
wa-hx
hs-ck
qo-hc
xm-hz
qu-xb
od-fz
yg-kn
ge-mj
rt-yi
nq-ek
iw-mh
pr-oq
tl-mw
kw-uc
wd-in
pn-ww
yk-gp
tm-sf
bf-yu
ou-gz
ss-hk
od-ox
mr-yr
zq-xp
zn-pd
tf-az
ah-ot
rz-sb
xg-wb
sv-vh
jg-je
fe-tx
ia-kg
ss-pu
qj-ws
sy-pp
et-ui
xq-tm
ok-bh
ye-bj
hc-gn
cv-kj
fq-yi
po-ky
ry-tj
oj-ce
tw-vj
ji-qp
fk-rz
aa-zx
wf-av
cr-jy
rc-gj
wn-bg
wg-ld
tm-tk
hr-wq
rk-di
lu-hv
rx-vi
wd-rz
od-gh
le-jd
ef-em
yq-pw
ws-bw
of-zw
zg-mk
ha-in
wa-yh
yz-su
ky-ki
rr-uw
kv-zx"""

result = solve_part2(test_input)
print(f"LAN party password: {result}")
```
